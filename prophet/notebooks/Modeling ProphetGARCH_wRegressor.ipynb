```python
import pandas as pd
import matplotlib.pyplot as plt
import utils.helpers as helpers
import joblib
from prophet import Prophet
```

```python
boroughs = ['Bronx','Brooklyn', 'EWR', 'Manhattan', 'Queens', 'Staten Island']
```

```python
for b in boroughs:
    print(f"Prophet w Regressor: {b}")
    
    df = pd.read_parquet(f'../data/{b} - all txns.parquet.gz')
    df = df.reset_index()
    df['timestamp_hour'] = pd.to_datetime(df['timestamp_hour'])
    df = df.sort_values(by = 'timestamp_hour')
    df = df.rename(columns = {'num_txns_All': 'y', 'timestamp_hour': 'ds'})

    df = df[['ds', 'y', 
                   'rolling_avg_w1-w4_lags', 
                   #'is_long_weekend',
                   # 'is_rush_hour', 
                   #'is_business_hour', 
                   #'lag_168'
            ]]

    df_hist = df[df['ds'] < '2024-08-01']
    df_test = df[df['ds']>= '2024-08-01']
    
    df_hist = df_hist.dropna()
    
    model = joblib.load(f'../models/{b}_Prophet_wRegressor.joblib')

    forecast = model.predict(df_test)
    __ = helpers.performance_metrics(df_test['y'], forecast['yhat'])
    
    #model_loc = f'../models/{b}_Prophet_wRegressor.joblib'
    #forecast_loc = f'../data/forecasts/{b}_Aug2024_forecasts_wRegressor.csv'
    
    #joblib.dump(model, model_loc)
    #forecast.to_csv(forecast_loc)
```

```python
from arch import arch_model
import numpy as np

# Extract residuals
forecast['residuals'] = df_test['y'].values - forecast['yhat'].values

# Step 3: Fit GARCH model to residuals
garch_model = arch_model(forecast['residuals'].dropna(), vol='Garch', p=1, q=1, rescale=False)
garch_fit = garch_model.fit(disp='off')

# # Step 4: Forecast residual volatility
garch_forecast = garch_fit.forecast(horizon=168)
volatility = garch_forecast.variance[-1:]  # Volatility forecast for next 5 periods

prophet_forecast = forecast['yhat'][-168:].values  # Trend + seasonality forecast
combined_forecast = prophet_forecast + np.sqrt(volatility) * np.random.normal(0, 1, 168)
```
